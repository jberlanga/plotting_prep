```python
# plotting_scratch--airmass
#
# A scratchpad for astroplan airmass plots.
#
# Authored by Jazmin Berlanga Medina.
# Summer 2015.
# 

import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time
%matplotlib inline

# Pull in input from file.
#times, airmasses = np.loadtxt("airmass_data", delimiter='\t', usecols=(0, 1), unpack=True)
airmasses = np.loadtxt("airmass_data", delimiter='\t', usecols=(1,), unpack=True)

# Set up time objects.
start_time = Time('2015-01-15T19:00:00')
end_time = Time('2015-01-16T04:00:00')

delta_t = end_time - start_time
times = start_time + delta_t*np.linspace(0.0, 1.0, len(airmasses))

# Set up axes & plot.
ax = plt.gca()
ax.plot_date(times.plot_date, airmasses, color='r', label='Sirius')
ax.plot_date(times.plot_date, 1.5*airmasses, color='k')
plt.gcf().autofmt_xdate()
plt.gca().invert_yaxis()

# Set labels, title, legend, etc.
ax.set_ylabel("Airmass")
ax.set_xlabel("Hour")
ax.set_title("Airmass vs Time - Sirius - 2015-01-15")
ax.legend(shadow=True)

plt.show()
```

## Subplot example

```python
# Subplot example

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()

```
