# astropy Time objects

astropy.time docs suggest use of [plot_date](http://astropy.readthedocs.org/en/latest/time/#time-format) [format](http://astropy.readthedocs.org/en/latest/api/astropy.time.TimePlotDate.html#astropy.time.TimePlotDate) when [plotting with matplotlib](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot_date).

[TimeDelta](http://astropy.readthedocs.org/en/latest/time/#time-deltas) class objects are only formatted in units of JD or seconds, so can't be displayed in certain formats (but note that Time objects are internally represented as JD). 

```python
# Using astropy.time docs 
# http://astropy.readthedocs.org/en/latest/time/

# Recommended use TimePlotDate with matplotlib?
# http://astropy.readthedocs.org/en/latest/api/astropy.time.TimePlotDate.html#astropy.time.TimePlotDate

import numpy as np
from astropy.time import Time
%matplotlib inline

times_list = ['1988-05-01T00:00:00', 
              '1998-05-01T00:00:00', 
              '2008-05-01T00:00:00', 
              '2018-05-01T00:00:00']

time_object = Time(times_list, format='isot')


print("\n")
print("String format:")
print(time_object)

print("\n")
print("Just the third time:") 
print(time_object[2])

print("\n")
print("Now in the JD format:")
time_jd = time_object.jd
print(time_jd)

print("\n")
print("In the tt scale:")
time_tt = time_object.tt
print(time_tt)

print("\n")
print("Can also get time differences")
print("(in TAI timescale, in units of days):")
delta_t = time_object[1] - time_object[0]
print(delta_t)
delta_t

print("\n")
print("We can also populate a Time object with multiple dates,")
print("starting with one date.")
print("Use format:")
print("date_list = start_date + (end_date - start_date)*np.linspace(0.0, 1.0, no_of_steps)")
new_time_object = time_object[0] + delta_t*np.linspace(0.0, 1.0, 12)
print(new_time_object)

#time_object[0] + delta_t*np.linspace(0.0, 2.0, 10)

# Using time objects with matplotlib
# (from TimePlotDate documentation).
import matplotlib.pyplot as plt
jyear = np.linspace(2000, 2001, 20)
t = Time(jyear, format='jyear', scale='utc')
plt.plot_date(t.plot_date, jyear)
plt.gcf().autofmt_xdate()  # orient date labels at a slant
plt.draw()
```
